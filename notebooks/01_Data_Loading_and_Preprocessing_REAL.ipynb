# Notebook 1: Data Loading and Preprocessing (with REAL DATA)

This notebook loads the actual data files downloaded and processed from public sources. It replaces the placeholder loaders with code that reads the real CSVs.

```
import pandas as pd
import os

# Define paths to the processed datasets
base_path = '/home/ubuntu/arabic_datasets_processed/'
padt_msa_path = os.path.join(base_path, 'PADT_MSA')
padic_path = os.path.join(base_path, 'PADIC_MultiDialect')

# Load the PADT MSA dataset
print('--- Loading PADT MSA ---')
try:
    df_train = pd.read_csv(os.path.join(padt_msa_path, 'train.csv'))
    df_dev = pd.read_csv(os.path.join(padt_msa_path, 'dev.csv'))
    df_test = pd.read_csv(os.path.join(padt_msa_path, 'test.csv'))
    print(f'PADT MSA loaded: Train={len(df_train)}, Dev={len(df_dev)}, Test={len(df_test)}')
    print('Sample sentence:', df_train['text'].iloc[0])
except Exception as e:
    print(f'Error loading PADT MSA: {e}')

# Load the PADIC Multi-Dialect dataset
print('
--- Loading PADIC Multi-Dialect ---')
try:
    df_padic = pd.read_csv(os.path.join(padic_path, 'padic_sentences.csv'))
    print(f'PADIC loaded: {len(df_padic)} parallel sentences')
    print('Sample parallel sentence:')
    print(df_padic.iloc[0])
except Exception as e:
    print(f'Error loading PADIC: {e}')
```
